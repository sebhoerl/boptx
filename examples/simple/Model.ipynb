```python
from model import ModeShareModel
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Run the model
model = ModeShareModel(
    constants = [-1.5, -2.0, -3.0],
    beta = -1.5, capacity = 0.5, travel_time = 1.0,
    blending = 1e-2
)

states = model.run()
```

```python
# Plot the states

plt.figure(dpi = 120, figsize = (6, 3))

for k in range(3):
    plt.plot(states[:,k], label = "Alternative %d" % k)

    
plt.grid()
plt.xlabel("Iteration")
plt.ylabel("Share")
plt.legend(loc = "best")
plt.ylim([0, 1])

plt.tight_layout()
plt.savefig("figures/states.png")
```
