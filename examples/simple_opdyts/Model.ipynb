```python
from model import ModeShareModel
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Run the model
model = ModeShareModel(
    constants = [-1.5, -2.0, -3.0],
    beta = -1.5, capacity = 0.5, travel_time = 1.0,
    blending = 1e-2
)

states1 = model.run(iterations = 100)
states2 = model.run(iterations = 100, initial_state = states1[-1])
states3 = model.run(iterations = 100, initial_state = states2[-1])
states4 = model.run(iterations = 100, initial_state = states3[-1])
```

```python
# Plot the states

plt.figure(dpi = 120, figsize = (6, 3))

for k, states in enumerate((states1, states2, states3, states4)):
    plt.plot(np.arange(99) + k * 100, states[:-1,:])

    
plt.grid()
plt.xlabel("Iteration")
plt.ylabel("Share")
plt.ylim([0, 1])

plt.tight_layout()
plt.savefig("figures/states.png")
```
